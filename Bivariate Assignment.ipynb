```python
dataset=dataset.select_dtypes(include='number')
dataset.cov()



###### dataset.cov() - returns value error specified as strings cannot be converted as covariance.
###### dataset contains non-numerical values
#####  include the select_dtypes to filter a DataFrame to keep only numeric columns
```

```python
##### Covariance ( degree/entrance test)
##### Covariance ( entrance / mba )

cov_value = dataset['degree_p'].cov(dataset['etest_p'])
print("Covariancebetween Degree and Entrance :", cov_value)

```

```python
##### Covariance value of 21.929 indicates that Degree / Entrance have a positive small difference
```

```python
import matplotlib.pyplot as plt

plt.scatter(dataset['degree_p'], dataset['etest_p'])
plt.xlabel('degree_p')
plt.ylabel('etest_p')
plt.title('Relationship between Degree and Entrance')
plt.show()
```

```python
cov_value = dataset['etest_p'].cov(dataset['mba_p'])
print("Covariance etween Entrance and MBA :", cov_value)

```

```python
##### Covariance value of 16.88 indicates that Entrance / MBA have a positive small difference
```

```python
import matplotlib.pyplot as plt

plt.scatter(dataset['etest_p'], dataset['mba_p'])
plt.xlabel('etest_p')
plt.ylabel('mba_p')
plt.title('Relationship between Entrance and MBA')
plt.show()
```
